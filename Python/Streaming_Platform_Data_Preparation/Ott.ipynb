```python
import pandas as pd

df = pd.read_excel("ott.xlsx", sheet_name="Sheet1")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df = df.reset_index(drop=True)   # reset index
df['ID'] = df.index + 1
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df = df.dropna(subset=['title'])
```

```python
df.isna().sum()
```

```python
df["age"] = df["age"].str.replace("+", "", regex=False)
```

```python
df['prime_video'] = df['prime_video'].astype(int)
```

```python
df['netflix'] = df['netflix'].astype(int)
```

```python
df['year'] = df['year'].astype(int)
df['disney+'] = df['disney+'].astype(int)
```

```python
df["age"] = df["age"].fillna("Unrated")
```

```python
df["rotten_tomatoes"] = df["rotten_tomatoes"].str.replace("/100", "", regex=False)
```

```python
df.head()
```

```python
df = df.dropna(subset=['rotten_tomatoes'])
```

```python
df["rotten_tomatoes"] = df["rotten_tomatoes"].astype(int)
```
